```python
df = pd.read_csv("/mnt/storage/grid/home/eric/hmm2bert/data_prep/labelled_datasets/puffinCaller_labelled_dataset_with8Functional_balanced.csv")
#df.set_index("sequence_ID")
#df.drop(columns = "Unnamed: 0", inplace=True)

df.info
```

```python
le = preprocessing.LabelEncoder()
le.fit(df['labels'])
```

```python
print(list(le.classes_))
```

```python
# prints a numbered list of the labels in order
t = list(le.classes_)
n = 0
for i in t:
    print(str(n) + ") " + i)
    n += 1
```

```python
new_labels = le.transform(df['labels'])
```

```python
new_labels_dict = {"sequence_ID": df['sequence_ID'], "sequences": df['sequences'], "labels": new_labels}

labelled_df = pd.DataFrame(new_labels_dict)

labelled_df = labelled_df.set_index("sequence_ID")
labelled_df.to_csv("/mnt/storage/grid/home/eric/hmm2bert/data_prep/encoded_labels_datasets/puffinCaller_encoded_dataset_with8Functional_balanced_labels.csv")
```
